## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
%matplotlib notebook

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
Data = pd.merge(mouse_metadata, study_results, on = 'Mouse ID', how = 'outer')
# Display the data table for preview
Data
```

```python
# Checking the number of mice.
Data['Mouse ID'].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
a = pd.DataFrame(Data.groupby('Mouse ID')['Timepoint'].value_counts()>1)

#duplicates
a.loc[a['Timepoint']==True]
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
Data.loc[Data['Mouse ID']== 'g989']
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleandata = Data.loc[Data['Mouse ID'] != 'g989']

#Data
cleandata
```

```python
# Checking the number of mice in the clean DataFrame.
cleandata['Mouse ID'].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
cgroup = cleandata.groupby('Drug Regimen')
cleandatamean = cgroup['Tumor Volume (mm3)'].mean()
cleandatamedian = cgroup['Tumor Volume (mm3)'].median()
cleandatavariance = cgroup['Tumor Volume (mm3)'].var()
cleandatasd = cgroup['Tumor Volume (mm3)'].std()
cleandatasem = cgroup['Tumor Volume (mm3)'].sem()
summary = pd.DataFrame({'Mean':cleandatamean, 
              'Median': cleandatamedian, 
              'Variance' : cleandatavariance, 
              'Standard Deviation': cleandatasd,
              'SEM' : cleandatasem
             })
summary
# This method is the most straighforward, creating multiple series and putting them all together at the end.

```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pandas. 
barone = pd.DataFrame(cgroup['Mouse ID'].nunique())
barone.plot(kind='bar',figsize= (6,3))
plt.xlabel('treatment')
plt.ylabel('number of mouse')
plt.title('mouse per treatment')
plt.tight_layout()
```
